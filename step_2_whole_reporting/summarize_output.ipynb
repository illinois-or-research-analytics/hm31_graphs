```python
import pandas as pd
import matplotlib.pyplot as plt

whole_stats = {}

for key, value in stats.items():
    if key == 'time':
        continue
        
    for sub_key, sub_value in value.items():
        if sub_key == 'PubmedArticle':
            continue
            
        if not sub_key in whole_stats:
            whole_stats[sub_key] = []
        
        whole_stats[sub_key].append(sub_value/value['PubmedArticle'])

plt.figure(figsize=(18, 12))

data_df = pd.DataFrame(whole_stats)

data_df.boxplot(column=['DateCompleted', 'ISSN', 'GrantList', 'MeshHeadingList', 'ChemicalList', 'PubDate'])

# Set the title and labels
plt.title("Box Plot of Each Attribute")
plt.ylabel("Percentage of presennce")
plt.xlabel("PubMed XML metadata statistics")

# Show the plot
plt.show()
    
```
